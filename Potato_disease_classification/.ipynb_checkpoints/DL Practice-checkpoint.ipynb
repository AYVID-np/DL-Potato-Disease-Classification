## Instead of using model.predict, write our own prediction function


```python
def sigmoid(x):
    import math
    
    return 1/(1+math.exp(-x))    

sigmoid(12)
```

```python
#y=coeff1*age + coeff2*affordability +intercept

def predcition_function(age , affordabilty):
    weighted_sum = coef[0]*age + coef[1]*affordabilty + intercept
    return sigmoid(weighted_sum)
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12,0,1]))
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1-15
    y_predicted_new = [max(i, epsilon) for i in y_predicted]
    y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gradient_descend(age, affordibility, y_true, epochs, loss_threshold):
    w1=w2=1
    bias=0
    rate=0.5
    n=len(age)
    print(epochs)
    for i in range(epochs):
        weighted_sum = w1*age + w2*affordibility + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)
        
        w1d = (1/n)*np.dot(age, (y_predicted-y_true))
        w2d = (1/n)*np.dot(affordibility, (y_predicted-y_true))
                     
        bias_d = np.mean(y_predicted-y_true)
                     
        w1 = w1 - rate*w1d
        w2 = w2 - rate*w2d
        bias = bias - rate*bias_d
        print(f'epochs: {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}')
                     
        if loss<= loss_threshold:
                     break
        
    return w1, w2, bias
```

## Writing custom Neural Network class from scratch
### Very Important

```python
# Aim
# customModel = myNN()
# customModel.fit(X_train_scaled, y_train, epochs=1000, loss_theshold=0.045)
# customModel.predict(X_test_scaled)

# should behave as keras predict function
```

```python
class myNN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
    
    def fit(self, X, y, epochs, loss_theshold):
        self.w1, self.w2, self.bias = self.gradient_descend(X['age'], X['affordibility'], y, epochs, loss_theshold)

    def predict(self, X_test):
        weighted_Sum = self.w1*X_test['age'] + self.w2*X_test['affordibility'] + self.bias
        return sigmoid_numpy(weighted_Sum)
    
    def gradient_descend(self, age, affordibility, y_true, epochs, loss_threshold):
        w1=w2=1
        bias=0
        rate=0.5
        n=len(age)
        print(epochs)
        for i in range(epochs):
            weighted_sum = w1*age + w2*affordibility + bias
            y_predicted = sigmoid_numpy(weighted_sum)
            loss = log_loss(y_true, y_predicted)

            w1d = (1/n)*np.dot(age, (y_predicted-y_true))
            w2d = (1/n)*np.dot(affordibility, (y_predicted-y_true))

            bias_d = np.mean(y_predicted-y_true)

            w1 = w1 - rate*w1d
            w2 = w2 - rate*w2d
            bias = bias - rate*bias_d
            
            if i%50 ==0:
                print(f'epochs: {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}')

            if loss<= loss_threshold:
                         break
        print(f' Best epochs: {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}')
        return w1, w2, bias
```

```python
customModel = myNN()
customModel.fit(X_train_scaled, y_train, epochs=500, loss_theshold=0.4631)

```
